```python
isbn10s = [] 
isbn13s = [] 
titles = []
authors = [] 
ranks = [] 
last_week_ranks = [] 
num_weeks_onlists = [] 
prices = []
descriptions = [] 
publishers = [] 
images = [] 
```

```python
end = time.time() + 3 

while time.time() <= end:  
    for book in books:
        
        try: isbn10 = book['primary_isbn10']
        except: isbn10 = None
            
        try: isbn13 = book['primary_isbn13']
        except: isbn13 = None
            
        try: title = book['title']
        except: title = None
            
        try: author = book['author']
        except: author = None
            
        try: rank = book['rank']
        except: rank = None
            
        try: last_week_rank = book['rank_last_week']
        except: last_week_rank = None
            
        try: num_weeks_onlist = book['weeks_on_list']
        except: num_weeks_onlist = None
            
        try: price = book['price']
        except: price = None
            
        try: description = book['description']
        except: description = None
            
        try: publisher = book['publisher']
        except: publisher = None
            
        try: image = book['book_image']
        except: image = None
  
        isbn10s.append(isbn10)
        isbn13s.append(isbn13)
        titles.append(title)
        authors.append(author)
        ranks.append(rank)
        last_week_ranks.append(last_week_rank)
        num_weeks_onlists.append(num_weeks_onlist)
        prices.append(price)
        descriptions.append(description)
        publishers.append(publisher)
        images.append(image)
       
```

```python
print(len(isbn10s))
print(len(isbn13s))
print(len(titles))
print(len(authors))
print(len(ranks))
print(len(last_week_ranks))
print(len(num_weeks_onlists))
print(len(prices))
print(len(descriptions))
print(len(publishers))
print(len(images))
```

```python
bookList = list(zip(titles, authors, isbn10s, 
                    isbn13s, ranks, last_week_ranks, num_weeks_onlists, 
                    prices, descriptions, publishers, images))

colnames = ['Title', 'Author', 'ISBN10', 'ISBN13', 'Rank', 'Rank of Last Week', 
            'Weeks on List', 'Price', 'Description', 'Publisher', 'Image']
```

```python
df = pd.DataFrame(bookList, columns=colnames)

df.head(10)
```
